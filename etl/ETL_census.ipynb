*Se eliminan las columnas actuales para dar paso a la fila [1] con el verdadero nombre de las columnas*

```python
nuevos_nombres = censo_edad.iloc[0]
censo_edad = censo_edad.iloc[1:].copy()
censo_edad.columns = nuevos_nombres

censo_edad.reset_index(drop=True, inplace=True)
```

- *Lista de columnas definitivas para el modelo*

```python
columnas_censo = [
    # IdentificaciÃ³n
    "Geographic Area Name",

    # Totales generales
    "Count!!SEX AND AGE!!Total population",
    "Count!!SEX AND AGE!!Male population",
    "Count!!SEX AND AGE!!Female population",

    # Edad media
    "Count!!MEDIAN AGE BY SEX!!Both sexes",

    # Hogares unipersonales
    "Count!!HOUSEHOLDS BY TYPE!!Total households!!Male householder, no spouse or partner present:!!Living alone",
    "Count!!HOUSEHOLDS BY TYPE!!Total households!!Female householder, no spouse or partner present:!!Living alone"
]

# Filtrar el DataFrame
census = censo_edad[columnas_censo].copy()
```

- *Renombrar columnas a nombres en espaÃ±ol simplificados*

```python
census.rename(columns={
    "Geographic Area Name": "localidad",

    # Totales
    "Count!!SEX AND AGE!!Total population": "poblacion_total",
    "Count!!SEX AND AGE!!Male population": "hombres_total",
    "Count!!SEX AND AGE!!Female population": "mujeres_total",


    # Edad media
    "Count!!MEDIAN AGE BY SEX!!Both sexes": "edad_media",

    # Hogares solos
    "Count!!HOUSEHOLDS BY TYPE!!Total households!!Male householder, no spouse or partner present:!!Living alone": "hombres_solos",
    "Count!!HOUSEHOLDS BY TYPE!!Total households!!Female householder, no spouse or partner present:!!Living alone": "mujeres_solas"
}, inplace=True)
```

```python
cols_to_convert = ['poblacion_total', 'hombres_total', 'mujeres_total', 
                   'edad_media', 'hombres_solos', 'mujeres_solas']

census[cols_to_convert] = census[cols_to_convert].replace(',', '', regex=True).astype(float)

# Verificar los tipos de datos
print(census.dtypes)
```
